```python
Weather = pd.read_csv("ChicagoWeather2011-2015.csv", parse_dates=['CST'])
Weather.head()
```

```python
print(Weather.count())
```

```python
Weather.rename(columns = {'CST':'KeyDate'}, inplace = True)
Weather['KeyDate'] = pd.to_datetime(Weather['KeyDate']).dt.date
Weather.head()
```

```python
Weather2 = Weather.set_index(Weather.KeyDate)
Weather2.head()
```

```python
print(Weather2.ix['2011-1-4'])
```

```python
Weather2[' Events'].fillna('Clear Skies!', inplace=True)
```

```python
Weather2[[' Events', 'KeyDate']].groupby([' Events']).count()
```

```python
Felonies = pd.read_csv("ChicagoCrime2015.csv", parse_dates=['Date'])
Felonies.head()
```

```python
FeloniesSince2015 = FeloniesSince2015.set_index(FeloniesSince2015.KeyDate)
FeloniesSince2015.head()
```

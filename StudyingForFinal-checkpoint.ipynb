```python
# import numpy and pandas, and DataFrame / Series
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import datetime

# Set some pandas options
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

# And some items for matplotlib
%matplotlib inline 
import matplotlib.pyplot as plt
# pd.options.display.mpl_style = 'default'

filename =  'june_airplane_data.csv'
# 'Chicago Crime 2011-2015.csv' # Chicago Weather Data 2011-2015.csv'   # 'june_airplane_data.csv' 'census.csv'
fieldOfInterest = "ArrDelay"  # age" # "Events" # Max TemperatureF"  # "DayOfWeek"  # "ArrDelay"
```

```python
df = pd.read_csv(filename, 
                 names=pd.read_html("http://stat-computing.org/dataexpo/2009/the-data.html")[1][1][1:])
```

```python
print(len(df.columns))
```

```python
df["ArrDelay"].plot(kind="hist", bins=60, title="Arrival Delays")
```

```python
df["Cancelled"].groupby(df["DayOfWeek"]).sum().plot(kind="bar")
```

```python
GMean = df.groupby(["UniqueCarrier","FlightNum"]).mean()
GMedian = df.groupby(["UniqueCarrier","FlightNum"]).median()
Diff = pd.DataFrame(GMean["ArrTime"] - GMedian["ArrTime"])
Diff["ArrTime"] = abs(Diff["ArrTime"])
print(Diff.sort_values("ArrTime", ascending=False))
```

```python
Diff2 = Diff.sort_values("ArrTime", ascending=True)
print(Diff2)

print(Diff2.dropna())
```
